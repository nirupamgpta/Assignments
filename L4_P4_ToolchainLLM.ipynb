```python
from pydantic import BaseModel, Field
from langchain_core.tools import Tool # Corrected import path for Tool
import json

# 2. Define a Pydantic model for the tool's input schema
class MongoDBQueryInput(BaseModel):
    query: dict = Field(description="A dictionary representing the MongoDB query document.")
    collection_name: str = Field(description="The name of the MongoDB collection to query.")

# 3. Create a Python function to execute the MongoDB query
def run_mongodb_query(query: dict, collection_name: str) -> str:
    """Executes a MongoDB find query on the specified collection and returns the results."""
    try:
        # Ensure the collection object is accessible. It was initialized in a previous step.
        # For this example, assuming 'db' is accessible from the previous cell.

        # Access the specified collection dynamically
        target_collection = db[collection_name]

        results = list(target_collection.find(query))
        if results:
            # Convert ObjectId to string for JSON serialization
            for doc in results:
                if '_id' in doc:
                    doc['_id'] = str(doc['_id'])
            return json.dumps(results, indent=2)
        else:
            return f"No documents found matching the query in collection '{collection_name}'."
    except Exception as e:
        return f"Error executing MongoDB query: {e}"

# 4. Instantiate the Tool object
mongodb_tool = Tool(
    name="mongodb_query_tool",
    func=run_mongodb_query,
    description="""Use this tool to query a MongoDB database. Input should be a JSON string with 'query' (a MongoDB query document as a dictionary) and 'collection_name' (the name of the collection).""",
    args_schema=MongoDBQueryInput
)

print("MongoDB Langchain Tool created successfully.")
print(f"Tool name: {mongodb_tool.name}")
print(f"Tool description: {mongodb_tool.description}")
```

```python
from langchain.agents import AgentExecutor, create_tool_calling_agent # Reverting to canonical import path
from langchain_core.prompts import ChatPromptTemplate

# tools must already be defined
tools = [mongodb_tool]

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful AI assistant that can query a MongoDB database. Use the 'mongodb_query_tool' for any database queries."),
        ("human", "{input}"),
        ("placeholder", "{agent_scratchpad}"),
    ]
)

agent = create_tool_calling_agent(llm, tools, prompt)

agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    verbose=True
)

print("LangChain Agent and AgentExecutor created successfully.")
```
